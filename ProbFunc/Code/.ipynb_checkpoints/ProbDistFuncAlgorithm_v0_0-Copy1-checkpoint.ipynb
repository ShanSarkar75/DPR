## Sampling from a PDF using Inverse Transform Sampling

**Given:** A function $f(x)$ defined on the domain $[a, b]$, representing an unnormalized or normalized probability density function (PDF).

---

### 1. Discretize the domain into $N$ points:

$$
x_i = a + (i-1)\Delta x, \quad i = 1, 2, \dots, N, \quad \Delta x = \frac{b - a}{N - 1}
$$

---

### 2. Normalize the PDF:

$$
f_i = \frac{f(x_i)}{\sum_{j=1}^N f(x_j)\, \Delta x}
\quad \text{so that} \quad
\sum_{i=1}^N f_i \Delta x = 1
$$

---

### 3. Compute the discrete cumulative distribution function (CDF):

$$
F_i = \sum_{k=1}^i f_k \Delta x,
\quad F_1 = f_1 \Delta x, \quad F_N = 1
$$

---

### 4. Define the inverse CDF using interpolation:

$$
F^{-1} : [0, 1] \to [a, b]
$$

$$
F^{-1}(u) \approx x_j \quad \forall\, u \in [0, 1], \quad \text{where} \quad F_j \leq u < F_{j+1}
$$

---

### 5. Generate samples:

- Draw $u_n \sim \mathcal{U}(0, 1)$ for $n = 1, 2, \dots, M$
- Map each $u_n$ to a sample:

$$
x_{\text{sample}, n} = F^{-1}(u_n)
$$


```python
"""
Generate samples from a given PDF using Inverse Transform Sampling.

Args:
    pdf: probability distribution
    x: 1D array of points in the domain [a, b]
    N: Number of samples to generate

Returns:
    samples: 1D array of sampled points from the PDF
"""

def InvTransformSampling(pdf, x, N=nSAMPLES):
    x=np.linspace(xMIN,xMAX,pdf.size)

    dx = np.diff(x)
    dx = np.append(dx, dx[-1])  # Handle last element spacing

    # Normalize PDF to ensure it integrates to 1
    pdf = pdf / np.sum(pdf * dx)

    # Compute the CDF
    cdf = np.cumsum(pdf * dx)
    cdf[-1] = 1.0  # Ensure numerical stability

    # Build the inverse CDF interpolator
    inv_cdf = interp1d(cdf, x, bounds_error=False, fill_value=(x[0], x[-1]))

    # Sample from uniform distribution and map via inverse CDF
    u = np.random.uniform(0, 1, N)
    samples = inv_cdf(u)    
    return samples
```

## Skew-Guassian Distribution with Mean $\mu$ and Standard Deviation $\sigma$

$$
f(x; \mu, \sigma, \alpha) = \frac{2}{\sigma} \, \phi(z) \, \Phi(\alpha z)
$$

Where:
- $z = \frac{x - \mu}{\sigma}$
- $\phi(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)$ is the standard normal PDF  
- $\Phi(x)$ is the standard normal CDF  
- $\mu$: mean 
- $\sigma$: standard deviation 
- $\alpha$: skewness parameter  
    - $\alpha = 0$: symmetric (normal)  
    - $\alpha > 0$: right-skewed  
    - $\alpha < 0$: left-skewed

**PDF of Skew-Guassian Distribution with Mean $\mu$ and Standard Deviation $\sigma$**
$$
\int_{-\infty}^{\infty} f(x; \mu, \sigma, \alpha) \, dx = 1
$$

**Case 1:** Symmetric (Normal) - $\alpha = 0$ | $\mu = 0$ | $\sigma = \sqrt{V}, \quad \text{where } V \sim \mathcal{U}(1, 5)$ |

**Case 2:** Right-skewed - $\alpha \sim \mathcal{U}(1, 5)$ (i.e., $\alpha > 0$) | $\mu = 0$ | $\sigma = \sqrt{V}, \quad \text{where } V \sim \mathcal{U}(1, 5)$ |

**Case 3:** Left-skewed - $\alpha \sim -\mathcal{U}(1, 5)$ (i.e., $\alpha < 0$) | $\mu = 0$ | $\sigma = \sqrt{V}, \quad \text{where } V \sim \mathcal{U}(1, 5)$ |


```python
from scipy.stats import norm
# Function to create Gaussian PDF with or without skew
def CreateGaussianPDF(x, mu, sigma, alpha):
    z = (x - mu) / sigma
    pdf = (2 / sigma) * norm.pdf(z) * norm.cdf(alpha * z)
    return pdf
```

## RetFunc: Piecewise Polynomial Regression for Probability Estimation (TensorFlow)

This function maps input tensor `x` to a probability tensor using two piecewise polynomial functions defined over [xMin, xMid] and [xMid, xMax], with coefficients `CoefLeft` and `CoefRight`, respectively.

### **Memory Breakdown** for `RetFunc` Function

Approximate memory estimate for 1D input of length \( W \) and polynomial order \( O \) (ignoring batch size), assuming float32 data type (4 bytes):

| Tensor        | Shape   | Number of elements | Memory (bytes)  | Description       |
| ------------- | --------| ------------------ | --------------- | ----------------- |
| **x**         | (W,)    | W                  | 4 \* W          | Input vector      |
| **powers**    | (O,)    | O                  | 4 \* O          | Powers vector     |
| **xExpand**   | (W, 1)  | W                  | 4 \* W          | Expanded input    |
| **xPow**      | (W, O)  | W \* O             | 4 \* W \* O     | Powers of input   |
| **CoefLeft**  | (1, O)  | O                  | 4 \* O          | Coefficients      |
| **CoefRight** | (1, O)  | O                  | 4 \* O          | Coefficients      |
| **yLeftRaw**  | (W,)    | W                  | 4 \* W          | Polynomial result |
| **yRightRaw** | (W,)    | W                  | 4 \* W          | Polynomial result |
| **IdxLeft**   | (W,)    | W                  | 4 \* W          | Mask              |
| **IdxRight**  | (W,)    | W                  | 4 \* W          | Mask              |
| **Prob**      | (W,)    | W                  | 4 \* W          | Final output      |

Total memory (approximate):

$$
\text{Memory} \approx 4 \times \big( 7W + 3 \times \text{O} + W \times \text{Order} \big) \quad \text{bytes}
$$

$$
\text{Memory}_{\text{KB}} = \frac{\text{Memory}}{1024} = \frac{4 \times (7W + 3O + W \times O)}{1024} \quad \text{KB}
$$

$$
\text{Memory}_{\text{MB}} = \frac{\text{Memory}}{1024^2} = \frac{4 \times (7W + 3O + W \times O)}{1024^2} \quad \text{MB}
$$

```python
# -----------------------------------------------------------------------------

"""
RetFunc: Piecewise Polynomial Regression for Probability Estimation (TensorFlow)

Computes piecewise polynomial regression-based probabilities over input tensor `x`, 
with separate polynomial coefficients applied on left and right intervals defined by `xMid`. 

Parameters:
- x: Input tensor of shape (WinSize),(Batch, WinSize) or (Batch, WinNo, WinSize)
- Order: Polynomial order (integer)
- xMid: Boundary point separating left and right intervals (scalar)
- CoefLeft: Polynomial coefficients for left interval, shape compatible with `Order`
- CoefRight: Polynomial coefficients for right interval, shape compatible with `Order`
- xMin, xMax: Minimum and maximum bounds for input values
- MinProb, MaxProb: Clipping bounds for output probabilities
- Auc: Normalization constant (area under curve)

Returns:
- Probabilities tensor of same shape as `x`.
"""

# -----------------------------------------------------------------------------
@tf.function#(reduce_retracing=True)
def RetFunc(x,                  # Input tensor (W),(B, W) or (B, C, W)
            Order,              # Polynomial order (integer)           
            xMid,               # Boundary point separating left and right intervals 
            CoefLeft,           # Polynomial coefficients for left interval - compatible with `Order`
            CoefRight,          # Polynomial coefficients for Right interval - compatible with `Order` 
            xMin,               # Minimum bound value for input x
            xMax,               # Maximum bound value for input x
            MinProb=1e-12,      # Minimum Clipping bound value for output probabilities
            MaxProb=1,          # Maximum Clipping bound value for output probabilities
            Auc=1):             # Normalization constant (Area Under Curve - PDF of Population)

    """
    x: Tensor of shape (WinSize), (Batch, WinSize) or (Batch, WinNo, WinSize)
    Returns: Probabilities of same shape
    """
    x = tf.cast(x, tf.float32)                     # shape: (Batch, WinSize) or (Batch, WinNo, WinSize)

    powers = tf.range(Order, dtype=tf.float32)     # shape: (Order,)
    xExpand = tf.expand_dims(x, axis=-1)           # shape (..., 1)
    xPow = tf.pow(xExpand, powers)                 # shape (..., Order)

    # Reshape coefficients for broadcasting
    ExpandShape = [1] * (len(x.shape)) + [Order]   # shape: [1, 1, ..., Order] matching xPow trailing dim
    CoefLeft = tf.reshape(CoefLeft, ExpandShape)   # shape: ExpandShape (broadcastable)
    CoefRight = tf.reshape(CoefRight, ExpandShape) # shape: ExpandShape (broadcastable)
    
    # Compute polynomials via einsum (more optimized on GPUs)
    #yLeftRaw = tf.reduce_sum(xPow * CoefLeft, axis=-1)
    yLeftRaw = tf.einsum('...o,...o->...', xPow, CoefLeft)    # shape: same as x without last dim
    # yRight = tf.reduce_sum(xPow * CoefRight, axis=-1)
    yRightRaw = tf.einsum('...o,...o->...', xPow, CoefRight)  # shape: same as x without last dim

    # Create masks
    IdxLeft = tf.cast(tf.logical_and(x >= xMin, x <= xMid), tf.float32) # shape: same as x
    IdxRight = tf.cast(tf.logical_and(x > xMid, x <= xMax), tf.float32) # shape: same as x
    
    # Apply masks and normalize
    Prob = (yLeftRaw * IdxLeft + yRightRaw * IdxRight) / Auc  # shape: same as x without last dim
    Prob = tf.clip_by_value(Prob, MinProb, MaxProb)           # shape: same as x without last dim
    return Prob
```

## Function Get_xTreshold():

This function calculates a threshold point on the x-axis where the slope of y with respect to x changes significantly, detecting either an upward or downward rapid change based on the `Up` flag. It uses the gradient and a scaled threshold factor to find where the slope crosses this threshold.

```python
"""
Detects the x-value where the slope of y(x) changes rapidly, 
based on a gradient threshold scaled by ThFactor.

Args:
    x (array-like or tf.Tensor): 1D input array of x-values.
    y (array-like or tf.Tensor): 1D input array of y-values.
    ThFactor (float): Scaling factor for gradient threshold detection. (Default 0.5%)
    Up (bool): If True, detects rapid upward slope change; 
               if False, detects rapid downward slope change.

Returns:
    float: The x-value where the slope crosses the threshold.
"""

def Get_xTreshold(x,y, ThFactor=0.01, Up=False):
    # Convert tensors to NumPy arrays (float32)
    if tf.is_tensor(x): x = x.numpy().astype(np.float32)
    if tf.is_tensor(y): y = y.numpy().astype(np.float32)
        
    #For Upward Direction Consider 1st Half For Downward Direction Consider 2nd Half 
    sIdx=0 if Up else 1 + x.size//2
    eIdx=x.size//2 if Up else x.size
    # Compute first derivative
    dy = np.gradient(y[sIdx:eIdx], x[sIdx:eIdx])
    # Thresholding: 
    # For Upward Direction - Find where slope starts increasing rapidly â†’ look where it becomes more positive
    # For Downward Direction - Find where slope stops decreasing rapidly â†’ look where it becomes less negative
    threshold = ThFactor * np.max(dy)  if Up else ThFactor * np.min(dy)

    idx = np.argmax(dy > threshold)
    x_change = x[sIdx:eIdx][idx]
    return x_change
#=====================================================================
## Test Code - For Evaluation / Debugging Purpose
#=====================================================================
if 1:
    fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
    #-----------------------------------
    # Case 1 - Upward Direction
    x = np.linspace(0, 5, 500)
    y = 1 / (1 + np.exp(-5*(x - 3)))  # sigmoid-like function with sharp rise near x=3
    x_Up= Get_xTreshold(x,y, ThFactor=0.005, Up=True)
    ax[0].plot(x, y, label='y(x)')
    ax[0].axvline(x_Up, color='r', linestyle='--', label=f"Sharp Change @ x={x_Up:.2f}")
    ax[0].legend()
    ax[0].set_ylabel("y")
    #-----------------------------------
    # Case 2 - Downward -  decreasing then flattening
    x = tf.linspace(0.0, 5.0, 500)
    y = 1 - tf.math.sigmoid(tf.exp(x))
    x_Dn= Get_xTreshold(x,y, ThFactor=0.005, Up=False)
    ax[1].plot(x, y, label='y(x)')
    ax[1].axvline(x_Dn, color='r', linestyle='--', label=f"Sharp Change @ x={x_Dn:.2f}")
    ax[1].legend()
    ax[1].set_ylabel("y")
    #-----------------------------------
    plt.tight_layout()
    plt.show()
```

## Creates a Piecewise Polynomial Function Approximating a PDF

This function creates a probability density function (PDF) approximation from a given dataset using either Kernel Density Estimation (KDE) or Histogram Density Estimation (HDE) on a trimmed subset of the data $\left(\text{within} \le \pm Th \cdot \sigma \right)$. The resulting PDF is approximated via separate polynomial fits (least squares) on the left and right of the main peak.  

### Steps:
    1. Flatten and cast the input dataset to float32.
    2. Compute mean and standard deviation.
    3. Define trimming bounds (xMin, xMax) or use Â±TimesStd * Ïƒ.
    4. Select subset within the bounds.
    5. Estimate density using KDE (Silverman bandwidth) or HDE.
    6. Sort and split density curve at maximum point.
    7. Identify bounds where density gradient drops below threshold.
    8. Fit separate polynomials to left and right intervals.
    9. Normalize area under the curve (AUC).
    10. Return either normalized (x, y) points or a callable PDF function.


```python
"""
-----------------------------------------------------------------------------
CreateFunc4PDF: Creates a Piecewise Polynomial Function Approximating a PDF
-----------------------------------------------------------------------------
Args:
    Superset (tf.Tensor or array-like): Input dataset.
    Points (int): Number of evaluation points for KDE/HDE.
    PolyOrder (int): Polynomial order for curve fitting.
    xMin, xMax (float or None): Optional explicit bounds.
    TimesStd (float): Range in units of standard deviation for trimming.
    UseKDE (bool): If True, use KDE; otherwise use HDE.
    RetProb (bool): If True, return normalized points; else return PDF function.
    Print (bool): If True, print debug info.
    USE_MAX_Y (bool): If True, split curve at global max; else at midpoint.

Returns:
    If RetProb=True: Tuple[tf.Tensor, tf.Tensor] â†’ (x_values, normalized_density)
    Else: Callable[[tf.Tensor], tf.Tensor] â†’ PDF function mapping x â†’ probability.

Possible Options:
    - Points: Any positive integer (e.g., 100, 200, 512)
    - PolyOrder:  Integer â‰¥ 1 (e.g., 3, 5, 7)
    - xMin, xMax:  None (auto-calculated) or float values for manual bounds
    - TimesStd:  Positive float (default=5 trims to Â±TimesStd Ã— Ïƒ)
    - UseKDE:
        * True â†’ Kernel Density Estimation
        * False â†’ Histogram Density Estimation
    - RetProb:
        * True â†’ Returns (x, normalized y)
        * False â†’ Returns callable PDF function
    - Print:
        * True â†’ Verbose debugging output
        * False â†’ Silent
    - USE_MAX_Y:
        * True â†’ Split at density peak
        * False â†’ Split at midpoint of X
Notes:
    - KDE bandwidth is computed using Silverman's rule.
    - Polynomial fitting is done separately on each side of the peak to improve fit accuracy.
    - The returned PDF function enforces clipping between MinProb and MaxProb.
"""
# -----------------------------------------------------------------------------
# Should pass the whole data with outlier
def CreateFunc4PDF(Superset, Points=nPDF, PolyOrder=5, xMin=None, xMax=None, TimesStd=PROB_FUNC_TH, 
                   UseKDE=True, RetProb=False, Print=False, USE_MAX_Y=True):
    ThFactor=0.005
    if Print: tf.print("0. Superset: ",Superset.shape )
    Superset = tf.cast(tf.reshape(Superset, [-1]), tf.float32)
    if Print: tf.print("1. Superset: ",Superset.shape )
    Epsilon = tf.constant(1e-12, tf.float32)
    if Print: tf.print("2. Epsilon: ",Epsilon )
        
    mean, std = tf.reduce_mean(Superset), tf.math.reduce_std(Superset)
    if Print: tf.print("3. mean: ",mean, "\tstd: ", std)
        
    xMinLeft = tf.cast(xMin, tf.float32) if xMin is not None else mean - TimesStd * std
    xMaxRight = tf.cast(xMax, tf.float32) if xMax is not None else mean + TimesStd * std
    if Print: tf.print("4. xMinLeft: ",xMinLeft, "\txMaxRight: ", xMaxRight)
        
    Mask99 = tf.math.logical_and(Superset > xMinLeft, Superset < xMaxRight)
    if Print: tf.print("5. Mask99: ",Mask99.shape )
        
    # Subset using boolean_mask
    Superset99 = tf.boolean_mask(Superset, Mask99)
    Count99=tf.reduce_sum(tf.cast(Mask99, tf.int32))
    if Print: tf.print("6. Superset99: ",Superset99.shape, "\tCount99: ",Count99)
    tf.print('Superset99 StdDev: ',tf.math.reduce_std(Superset99))
    
    if UseKDE: X, Y = KDE(X=Superset99, Points=Points) # Compute BW using Silverman's rule
    else: X, Y,_ = HDE(S=Superset99, nBins=None, Epsilon=Epsilon,fSizeMin=3, fSigma=2.5)
    if Print: tf.print("7. X: ",X.shape,"  Y: ",Y.shape)
    #fig, axs = plt.subplots(1, 1, figsize=(9, 9)); axs.plot(X.numpy(), Y.numpy())
        
    SortedIdx = tf.argsort(X)
    if Print: tf.print("8. SortedIdx: ", SortedIdx.shape)
    xSorted, ySorted = tf.gather(X, SortedIdx), tf.gather(Y, SortedIdx)
    if Print: tf.print("9. xSorted: ",xSorted.shape, "\tySorted: ",ySorted.shape)
        
    MaxY = tf.reduce_max(ySorted)
    MidIdx = tf.argmax(ySorted) if USE_MAX_Y else Points // 2
    xAtMid = xSorted[MidIdx];# yAtMid=ySorted[MidIdx]
    if Print:tf.print("10. MaxY:", MaxY, "\tMidIdx:", MidIdx, "\txAtMid:", xAtMid)
        
    UpX, DnX = xSorted[:MidIdx], xSorted[MidIdx:]
    if Print: tf.print("11. UpX: ",UpX.shape, "\tDnX: ",DnX.shape)    
    UpY, DnY = ySorted[:MidIdx], ySorted[MidIdx:]
    if Print: tf.print("12. UpY: ",UpY.shape, "\tDnY: ",DnY.shape)
        
    # Update xMinLeft to x Point before which Gradient of y is very low
    xUpTh=Get_xTreshold(UpX,UpY, ThFactor, Up=True)
    if xUpTh > xMinLeft: xMinLeft=xUpTh
    if Print: tf.print("13. xUpTh: ",xUpTh, "\txMinLeft: ",xMinLeft )
        
    # Update xMaxRight to x Point after which Gradient of y is very low
    xDnTh=Get_xTreshold(DnX,DnY, ThFactor, Up=False)
    if xDnTh < xMaxRight: xMaxRight=xDnTh
    if Print: tf.print("14. xDnTh: ",xDnTh, "\txMaxRight: ",xMaxRight )
        
    xOrder = tf.stack([tf.pow(xSorted, i) for i in range(PolyOrder)], axis=1)
    if Print: tf.print("15. xOrder: ", xOrder.shape)
        
    UpX, DnX = xOrder[:MidIdx], xOrder[MidIdx:]
    if Print: tf.print("16. UpX: ",UpX.shape, "\tDnX: ",DnX.shape) 
    ExUpY, ExDnY = tf.expand_dims(ySorted[:MidIdx], axis=1), tf.expand_dims(ySorted[MidIdx:], axis=1)
    if Print: tf.print("17. ExUpY: ",ExUpY.shape, "\tExDnY: ",ExDnY.shape) 
        
    CoefUp = tf.linalg.lstsq(UpX, ExUpY)
    CoefDn = tf.linalg.lstsq(DnX, ExDnY)
    if Print: tf.print("18. CoefUp: ",CoefUp.shape, "\tCoefDn: ",CoefDn.shape) 
    tf.print("18. CoefUp: ",CoefUp, "\tCoefDn: ",CoefDn)   
    
    xUp = tf.linspace(xSorted[0], xSorted[MidIdx-1], Points)
    xDn = tf.linspace(xSorted[MidIdx], xSorted[-1], Points)
    if Print: tf.print("19. xUp: ",xUp.shape, "\txDn: ",xDn.shape) 
        
    yUp = tf.add_n([CoefUp[i] * tf.pow(xUp, i) for i in range(PolyOrder)])
    yDn = tf.add_n([CoefDn[i] * tf.pow(xDn, i) for i in range(PolyOrder)])
    if Print: tf.print("20. xUp: ",xUp.shape, "\txDn: ",xDn.shape) 
        
    xFit = tf.concat([xDn, xUp], axis=0)
    yFit = tf.concat([yDn, yUp], axis=0)
    yFit = tf.clip_by_value(yFit, Epsilon, MaxY)
    if Print: 
        tf.print("21. xFit: ",xFit.shape, "\tyFit: ",yFit.shape) 
        tf.print("22. Min(yFit):", tf.reduce_min(yFit), "\tMax(yFit):", tf.reduce_max(yFit))

        
    SortedIdx = tf.argsort(xFit)
    if Print: tf.print("23. SortedIdx: ",SortedIdx.shape)
        
    xFitSorted, yFitSorted = tf.gather(xFit, SortedIdx), tf.gather(yFit, SortedIdx)
    if Print: tf.print("24. xFitSorted: ",xFitSorted.shape, "\tyFitSorted: ",yFitSorted.shape) 
        
    dx = xFitSorted[1:] - xFitSorted[:-1]
    Av_dy = (yFitSorted[1:] + yFitSorted[:-1]) / 2.0
    BaseAUC = tf.reduce_sum(dx * Av_dy)
    if Print: tf.print("25. dx: ",dx.shape, "\tAv_dy: ",Av_dy.shape, "\tBaseAUC: ",BaseAUC) 
    

    if RetProb:
        return xFitSorted, yFitSorted/BaseAUC
    else:
        # Return a Polynomila Regression Function with Pre-trained coefficients and settings
        return lambda x: RetFunc(x,                  # Input tensor (W),(B, W) or (B, C, W)
                                 Order=PolyOrder,    # Polynomial order (integer)           
                                 xMid=xAtMid,        # Boundary point separating left and right intervals 
                                 CoefLeft=CoefUp,    # Polynomial coefficients for left interval
                                 CoefRight=CoefDn,   # Polynomial coefficients for Right interval  
                                 xMin=xMinLeft,      # Minimum bound value for input x
                                 xMax=xMaxRight,     # Maximum bound value for input x
                                 MinProb=Epsilon,    # Minimum Clipping bound value for output probabilities
                                 MaxProb=MaxY,       # Maximum Clipping bound value for output probabilities
                                 Auc=BaseAUC)        # Normalization constant (Area Under Curve - PDF of Population)
                                 #Print=True)
```

```python
# -----------------------------------------------------------------------------
# ComputeAUC: Estimates the area under a 1D curve using the trapezoidal rule.
#
# Inputs:
#   x : 1D tensor of x-values (must be sorted).
#   P : 1D tensor of corresponding probability or density values.
#
# Computes dx between consecutive x-values and integrates by averaging adjacent
# P values. Useful for verifying if a PDF integrates to ~1.
#
# Returns:
#   auc : Scalar tensor representing the estimated area under the curve.
# -----------------------------------------------------------------------------

def ComputeAUC(x, P):
    x = tf.cast(x, tf.float32)
    P = tf.cast(P, tf.float32)
    
    dx = x[1:] - x[:-1]
    avg_P = (P[1:] + P[:-1]) / 2.0
    auc = tf.reduce_sum(dx * avg_P)
    return auc
```

```python
# -----------------------------------------------------------------------------
# ComputeJSD: Computes the Jensen-Shannon Divergence (JSD) between two 
# probability distributions P and Q using TensorFlow operations.
#
# - Ensures numerical stability by clipping values to avoid log(0).
# - Computes the average distribution M = 0.5*(P + Q).
# - Uses the KL divergence formula to calculate JSD:
#       JSD(P || Q) = 0.5 * KL(P || M) + 0.5 * KL(Q || M)
# - Returns scalar JSD value; optionally prints intermediate shapes and values.
# - Supports both float32 and float64 precision.
# -----------------------------------------------------------------------------


def ComputeJSD(P,Q,epsilon=1e-12,PrintFlag = False):
    # Convert both tensors to the same dtype (float32 or float64)
    dtype = tf.float32  # or tf.float64 depending on your preference
    P = tf.cast(P, dtype)
    Q = tf.cast(Q, dtype)
    
    
    #  Clip to avoid log(0)
    P = tf.clip_by_value(P, epsilon, 1.0)
    Q = tf.clip_by_value(Q, epsilon, 1.0)
    tf.print("P shape:", tf.shape(P), " Q shape:", tf.shape(Q))
    # Average distribution
    M = 0.5 * (P + Q)
    M = tf.clip_by_value(M, epsilon, 1.0)

    # KL divergences
    kl_P = tf.reduce_sum(P * tf.math.log(P / M))
    kl_Q = tf.reduce_sum(Q * tf.math.log(Q / M))

    # Jensenâ€“Shannon divergence
    jsd = 0.5 * (kl_P + kl_Q)  

    if PrintFlag:
        tf.print("P shape:", tf.shape(P), " Q shape:", tf.shape(Q), " M shape:", tf.shape(M))
        tf.print("kl_P :", kl_P, " kl_Q:", kl_Q)
        tf.print("JSD :", jsd)

    return jsd 

if 0: ComputeJSD(pdf,pdfFunc,epsilon=1e-12,PrintFlag = True)
```

```python
# UpdateDf updates an existing row in the DataFrame if a match is found for
# both 'Distribution' and 'Function'; otherwise, it appends a new row.
#
# UpdateDf(df, distribution, function, memory_kb, training_time, inference_time, auc, js_divergence)
# 
# Args:
#   df (pd.DataFrame): The input DataFrame with performance metrics.
#   distribution (str): Name of the distribution type (e.g., 'KDE').
#   function (str): Name of the function or estimator used.
#   memory_kb (float): Memory usage in kilobytes.
#   training_time (float): Time taken for training.
#   inference_time (float): Time taken for inference/prediction.
#   auc (float): Area Under the Curve (AUC) metric.
#   js_divergence (float): Jensen-Shannon divergence value.
#
# Returns:
#   pd.DataFrame: Updated DataFrame with the new or modified row.

def UpdateDf(df, distribution, function, training_time, inference_time, auc, js_divergence): #memory_kb, 
    # Check if a row with the given distribution and function already exists
    mask = (df['Distribution'] == distribution) & (df['Function'] == function)

    if mask.any():
        # Update the existing row
        df.loc[mask, ['Training', 'Inference', 'AUC', 'JS-Divergence']] = [
            training_time, inference_time, auc, js_divergence
        ]
    else:
        # Add a new row
        new_row = {
            'Distribution': distribution,
            'Function': function,
            #'Memory(KB)': float(memory_kb),
            'Training': training_time,
            'Inference': inference_time,
            'AUC': float(auc),
            'JS-Divergence': float(js_divergence)
        }
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

    return df

```

## M-Wright Function

The M-Wright function is a special function that appears in the study of fractional calculus, anomalous diffusion, and time-fractional differential equations. 

$M_\nu(x) = \sum_{j=0}^\infty \frac{(-x)^j}{j! \, \Gamma(-\nu j + 1 - \nu)}$

**Where:**

- $\nu \in (0, 1)$: shape parameter (controls tail heaviness and spread)
- $x \in \mathbb{R}$: real-valued input (evaluation point)
- $j \in \mathbb{N}_0$: summation index (non-negative integers, i.e., $j = 0, 1, 2, \dots$)
- $\Gamma(\cdot)$: Gamma function (generalizes factorial, where $\Gamma(n) = (n - 1)!$ for $n \in \mathbb{N}$)

For positive integers $n$, the Gamma function $\Gamma$ satisfies:  $\Gamma(n) = (n - 1)!$

    

```python
from scipy.special import gamma

def M_Wright(x, v, J=50):
    """
    Compute the M-Wright function M_nu(x) using a truncated series expansion.
    Handles potential NaNs by skipping invalid terms.
    
    Assumes x >= 0 and 0 < v < 1.
    """
    x = np.asarray(x, dtype=np.float32)
    M = np.zeros_like(x)

    for j in range(int(J)):
        try:
            denom = np.math.factorial(j) * gamma(-v * j + 1 - v)
            if np.isnan(denom) or np.isinf(denom):
                continue  # Skip term if invalid
            term = ((-x) ** j) / denom
            M += term
        except (OverflowError, ZeroDivisionError, ValueError, FloatingPointError):
            continue  # Gracefully skip problematic term

    # Final safety check: replace any resulting NaNs with 0
    M = np.nan_to_num(M, nan=0.0, posinf=0.0, neginf=0.0)
    return M


#-------------------

```

## M-Wright Function Using Gamma Function Reflection Formula:
$$
\Gamma(1 - \tau) \, \Gamma(\tau) = \frac{\pi}{\sin(\pi \tau)}, \quad \text{for } \tau \notin \mathbb{Z}
$$

If we define:
$$
\tau = \nu j + \nu \quad \Rightarrow \quad -\nu j + 1 - \nu = 1 - (\nu j + \nu) = 1 - \tau
$$

Then:
$$
\Gamma(-\nu j + 1 - \nu) = \Gamma(1 - \tau) = \frac{\pi}{\Gamma(\tau)\sin(\pi \tau)}
$$

---

#### Substituting into the M-Wright Series:

**Original Series**:
$$
M_\nu(x) = \sum_{j=0}^{\infty} \frac{(-x)^j}{j! \, \Gamma(-\nu j + 1 - \nu)}
$$

**Using the reflection formula**:
$$
M_\nu(x) = \sum_{j=0}^{\infty} \frac{(-x)^j \, \Gamma(\nu j + \nu) \, \sin(\pi (\nu j + \nu))}{j! \, \pi}
$$

---

#### Final Expression:
$$
M_\nu(x) = \frac{1}{\pi} \sum_{j=0}^{\infty} \frac{(-x)^j \, \Gamma(\nu j + \nu) \, \sin(\pi (\nu j + \nu))}{j!}
$$

---

#### Benefit:
This form avoids the singularities of the Gamma function in the original denominator and may improve numerical stability â€” especially when
$-\nu j + 1 - \nu$ becomes negative.


```python
import numpy as np
from scipy.special import gamma

def M_Wright_Reflect(x, v, J=50):
    """
    Compute the M-Wright function M_v(x) using the reflection formula
    for improved numerical stability.
    
    Parameters:
    - x : array-like or scalar, non-negative input
    - v : float, 0 < v < 1
    - J : int, number of terms in the series (default 50)
    
    Returns:
    - M : array-like, same shape as x
    """
    x = np.asarray(x, dtype=np.float64)
    M = np.zeros_like(x)

    for j in range(int(J)):
        tau = v * j + v
        numerator = (-x)**j * gamma(tau) * np.sin(np.pi * tau)
        denominator = np.math.factorial(j) * np.pi
        coeff = numerator / denominator
        M += coeff

    return M
if 1:
    # Test
    v = 0.75
    J=100
    x = np.linspace(0, 5, nPDF)
    mw_x = M_Wright(x, v, J)#, Print=True)
    nan_count = np.sum(np.isnan(mw_x))
    print(f'Number of NaN values in mw_x: {nan_count}')
    
    
    mwR_x = M_Wright_Reflect(x, v, J)#, Print=True)
    nan_count = np.sum(np.isnan(mwR_x))
    print(f'Number of NaN values in mwR_x: {nan_count}')
    # Plot
    plt.plot(x, mw_x, c='r', ls=':', lw=2)
    plt.plot(x, mwR_x, c='b',ls='--', lw=1)
    plt.title(f'M-Wright Function ($\\nu$={v}, J={J})')
    plt.xlabel('x')
    plt.ylabel(f'$M_\\nu(x)$')
    plt.grid(True)
    plt.show()
    
```

## Asymmetric M-Wright Type I Distribution: $\mathrm{AWM}_\nu^{I}(\lambda)$

A random variable $X$ follows the **Asymmetric M-Wright Type I distribution** if its probability density function is given by:

$$
f_\nu(x; \lambda) = M_\nu(|x|) \cdot \mathfrak{M}_\nu(\lambda x), \quad 0 \leq \nu < 1
$$

Here,

- $M_\nu(|x|)$: This is the M-Wright function, which acts as a PDF kernel centered at zero. It is non-negative, integrable, and has exponential-type decay.

- $\mathfrak{M}_\nu(\lambda x)$: This is the cumulative distribution function (CDF) associated with the symmetric M-Wright PDF.

$$
\mathfrak{M}_\nu(y) = \frac{1}{2} \left[ 1 + \operatorname{sgn}(y) \left( 1 - \sum_{j=0}^\infty \frac{(-|y|)^j}{j! \, \Gamma(1 - \nu j)} \right) \right]
$$


This form introduces asymmetry by modulating the symmetric PDF $M_\nu(|x|)$ with its CDF $\mathfrak{M}_\nu(\lambda x)$. The parameter $\lambda$ governs the asymmetry and skewness.

## Asymmetric M-Wright Distribution Type II ( $\mathrm{AMW}_\nu^{\mathrm{II}}(\alpha)$ ) :


A random variable $X$ follows the **Asymmetric M-Wright Type II distribution** if its probability density function is given by:

$$
f_\nu(x; \alpha) =\frac{\alpha}{(1 + \alpha^2)}
\begin{cases}
M_\nu(\alpha x), & \text{if } x \geq 0 \\\\
M_\nu(-\frac{x}{ \alpha}), & \text{if } x < 0
\end{cases}
$$

where:
- $M_\nu(x)$ is the **M-Wright function** (PDF kernel),
- $0 \le \nu < 1$ is the shape parameter,
- $\alpha > 0$ is the asymmetry parameter.



```python
# Function to create Asymmetric M-Wright Type II Distribution 
def Create_AWM_II_PDF(x, v, alpha, J=50):
    """
    Compute the Asymmetric M-Wright Type II PDF:
    
        f_v(x; alpha) = (alpha / (1 + alpha^2)) * 
                        M_v(alpha * x) if x >= 0,
                        M_v(-x / alpha) if x < 0

    Parameters:
    - x : array-like or scalar input
    - v : fractional parameter (0 <= v < 1)
    - alpha : asymmetry parameter (alpha > 0)
    - J : number of series terms in M_Wright_Reflect

    Returns:
    - f : PDF values, same shape as x
    """
    x = np.asarray(x, dtype=np.float64)
    f = np.zeros_like(x)

    norm_const = alpha / (1 + alpha**2)

    # For x >= 0
    mask_pos = x >= 0
    f[mask_pos] = norm_const * M_Wright_Reflect(alpha * x[mask_pos], v, J)

    # For x < 0
    mask_neg = ~mask_pos
    f[mask_neg] = norm_const * M_Wright_Reflect(-x[mask_neg] / alpha, v, J)

    return f


if 1:
    # Test
    v = 0.25
    J=100
    Alpha=2.0 # 0.3-0.7 or 1.3 - 1.7
    x = np.linspace(xMIN, xMAX, nPDF)
    xPos = x[np.where(x >= 0)[0]]; y=xPos*Alpha
    M_AlphaPosX = M_Wright_Reflect(y, v, J)
    nan_count = np.sum(np.isnan(M_AlphaPosX))
    print(f'Number of NaN values in M_AlphaPosX: {nan_count}')
    
    xNeg = x[np.where(x < 0)[0]]; y=-xNeg/Alpha
    M_AlphaNegX = M_Wright_Reflect(y, v, J)
    nan_count = np.sum(np.isnan(M_AlphaNegX))
    print(f'Number of NaN values in M_AlphaNegX: {nan_count}')    
    M_AlphaAllX = np.concatenate([M_AlphaNegX, M_AlphaPosX])


    pdf = Create_AWM_II_PDF(x, v, Alpha, J)
    nan_count = np.sum(np.isnan(pdf))
    print(f'Number of NaN values in pdf: {nan_count}') 
    print(f"PDF AUC: {ComputeAUC(x, pdf):.3f}")
    # Plot
    plt.plot(x, M_AlphaAllX, c='b', ls=':', lw=2,label=r'$M_\nu(y)$')
    plt.plot(x, pdf, c='r',ls='--', lw=1, label=r'$f_\nu(x; \alpha)$')
    
    plt.title(f'M-Wright Function ($\\nu$={v}, $\\alpha$={Alpha}, J={J})')
    plt.xlabel('x')
    plt.ylabel(r'$f_\nu(x; \alpha)$  and  $M_\nu(y)$')
    plt.legend()
    plt.grid(True)
    plt.show()

    

    
```

# Asymmetric Laplace PDF

The **Asymmetric Laplace Distribution (ALD)** is a continuous distribution extending the classic Laplace by allowing different decay rates on each side of its central location. Hereâ€™s the mathematical form:

\\[
X \sim \text{AL}(m, \lambda, \kappa),
\\]  
where:

- \( $m$ \): location parameter  
- \( $\lambda > 0 $\): scale  
- \( $\kappa > 0 $\): asymmetry factor

The **probability density function (PDF)** is:

\\[
f(x \mid m, \lambda, \kappa) =
\frac{\lambda}{\kappa + \frac{1}{\kappa}}
\begin{cases}
\exp\left( \frac{\lambda}{\kappa}(x - m) \right), & x < m \\\\
\exp\left( -\lambda \kappa (x - m) \right), & x \ge m
\end{cases}
\\]

When \( $\kappa = 1 $\), this reduces to the symmetric Laplace distribution.

Kotz, S., Kozubowski, T.J., PodgÃ³rski, K. (2001). Asymmetric Laplace Distributions. In: The Laplace Distribution and Generalizations. BirkhÃ¤user, Boston, MA. https://doi.org/10.1007/978-1-4612-0173-1_3


```python
# Define the PDF of the Asymmetric Laplace Distribution
def AsymmetricLaplacePDF(x, m, lamda, kappa):
    norm_const = lamda / (kappa + 1/kappa)
    pdf = np.where(
        x < m,
        norm_const * np.exp((lamda / kappa) * (x - m)),
        norm_const * np.exp((-lamda * kappa) * (x - m))
    )
    return pdf

# Inverse transform sampling function
def SampleAsymmetricLaplace(m, lambda_, kappa, size):
    u = np.random.uniform(0, 1, size)
    p = kappa**2 / (1 + kappa**2)
    samples = np.where(
        u < p,
        m + kappa / lambda_ * np.log(u / p),
        m - 1 / (lambda_ * kappa) * np.log((1 - u) / (1 - p))
    )
    return samples

if 1: 
    # Parameters for the Asymmetric Laplace Distribution
    m = 0.2       # location
    lamda = 2 # scale
    kappa = 2.5 # asymmetry
    N=nPDF
    

    # Generate x values and corresponding PDF values
    x_vals = np.linspace(xMIN, xMAX, nPDF)
    pdf_vals = AsymmetricLaplacePDF(x_vals, m, lamda, kappa)

    # Generate the dataset X^
    X_hat = SampleAsymmetricLaplace(m, lamda, kappa, 100000)



    # Plot the distribution
    plt.figure(figsize=(8, 4))
    plt.hist(X_hat, bins=100, density=True, alpha=0.6, label="Sampled Histogram")
    plt.plot(x_vals, pdf_vals, label=f"m={m}, Î»={lamda}, Îº={kappa}")
    plt.title("Asymmetric Laplace Distribution")
    plt.xlabel("x")
    plt.ylabel("PDF")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Compute area under the PDF using Simpson's rule
    area = np.trapz(pdf_vals, x_vals)
    print(area)
```


#### Sampling Formula (Transformation-based) 

To simulate $x \sim \text{SkewNormal}(\mu, \sigma, \alpha)$:

$$
x = \mu + \sigma \left( \delta \cdot |Z_0| + \sqrt{1 - \delta^2} \cdot Z_1 \right)
$$

Where $Z_0, Z_1 \sim \mathcal{N}(0,1)$ are independent standard normal variables, and

$$
\delta = \frac{\alpha}{\sqrt{1 + \alpha^2}}
$$

## Sampling from a PDF using Inverse Transform Sampling

**Given:**  A function ${f}(x)$ is the target PDF on domain $[a, b]$ 

#### Steps:

**Discretization of the domain into $N$ points:**

$$
x_i = a + (i-1)\Delta x, \quad i = 1, 2, \dots, N, \quad \Delta x = \frac{b - a}{N-1}
$$

**Normalize the PDF**

$$
f_i = \frac{{f}(x_i)}{\sum_{j=1}^N {f}(x_j) \Delta x} \quad \text{such that} \quad \sum_{i=1}^N f_i \Delta x = 1
$$

**Compute the discrete cumulative distribution function (CDF)**

$$
F_i = \sum_{k=1}^i f_k \Delta x, \quad F_1 = f_1 \Delta x, \quad F_N = 1
$$

**Build an interpolation function for the inverse CDF**

$$
F^{-1} : [0, 1] \to [a, b]
$$

$$
F^{-1}(u) \approx x_j \quad \forall \quad u \in [0,1], \quad \text{where} \quad F_j \leq u < F_{j+1}
$$

**Generate samples**

- Draw uniform random numbers: $u_n \sim \mathcal{U}(0, 1), \quad n = 1, 2, \dots, N$

- Map each $u_n$ to a sample: $x_{\text{sample}, n} = F^{-1}(u_n)$


## Sampling from a PDF using Inverse Transform Sampling

**Given:**

- A function $\tilde{f}(x)$ proportional to the target PDF on domain $[a, b]$ (e.g., $a = -10, b = 10$).
- A discretization of the domain into $N$ points:

$$
x_i = a + (i-1)\Delta x, \quad i = 1, 2, \dots, N, \quad \Delta x = \frac{b - a}{N-1}
$$

---

**Steps:**

1. **Normalize the PDF**

$$
f_i = \frac{\tilde{f}(x_i)}{\sum_{j=1}^N \tilde{f}(x_j) \Delta x} \quad \text{such that} \quad \sum_{i=1}^N f_i \Delta x = 1
$$

2. **Compute the discrete cumulative distribution function (CDF)**

$$
F_i = \sum_{k=1}^i f_k \Delta x, \quad F_1 = f_1 \Delta x, \quad F_N = 1
$$

3. **Build an interpolation function for the inverse CDF**

Define a function

$$
F^{-1} : [0, 1] \to [a, b]
$$

such that for any $u \in [0,1]$,

$$
F^{-1}(u) \approx x_j \quad \text{where} \quad F_j \leq u < F_{j+1}
$$

4. **Generate samples**

- Draw uniform random numbers

$$
u_m \sim \mathcal{U}(0, 1), \quad m = 1, 2, \dots, M
$$

- Map each $u_m$ to a sample

$$
x_{\text{sample}, m} = F^{-1}(u_m)
$$

---

**Summary:**

$$
\boxed{
\begin{cases}
f_i = \frac{\tilde{f}(x_i)}{\sum_{j=1}^N \tilde{f}(x_j) \Delta x} \\
F_i = \sum_{k=1}^i f_k \Delta x \\
x_{\text{sample}, m} = F^{-1}(u_m), \quad u_m \sim \mathcal{U}(0,1), \quad m=1,\ldots,M
\end{cases}
}

```python
# 1. M-Wright function
def M_wright(z, nu, terms=50):
    z = np.asarray(z, dtype=np.float64)
    result = np.zeros_like(z)
    for k in range(terms):
        coeff = ((-z) ** k) / (np.math.factorial(k) * gamma(-nu * k + 1 - nu))
        result += coeff
    return result

# 2. Asymmetric PDF
def asymmetric_M_wright_pdf(x, nu, lam, terms=50):
    abs_x = np.abs(x)
    return M_wright(abs_x, nu, terms) * M_wright(lam * x, nu, terms)

# 3. Compute PDF and normalize
x = np.linspace(-10, 10, 800)
nu = 0.3
lam = 0.8
pdf = asymmetric_M_wright_pdf(x, nu, lam, terms=60)

dx = x[1] - x[0]
pdf /= np.sum(pdf * dx)  # Normalize so that integral â‰ˆ 1

# 4. Compute CDF
cdf = np.cumsum(pdf) * dx
cdf /= cdf[-1]  # Normalize to 1

# 5. Inverse transform sampling
inv_cdf = interp1d(cdf, x, bounds_error=False, fill_value=(x[0], x[-1]))

# Generate samples
num_samples = 10000
u = np.random.rand(num_samples)
samples = inv_cdf(u)

# 6. Plot PDF and Histogram
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label='Target PDF', linewidth=2)
plt.hist(samples, bins=80, density=True, alpha=0.6, label='Sampled Histogram')
plt.title('Samples from Asymmetric M-Wright Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


from scipy.integrate import simps

area_simps = simps(pdf, x)
print(f"Area under the curve using Simpsonâ€™s rule: {area_simps:.6f}")
```

```python
def ComputeSkewness(data):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    skewness = (np.sum((data - mean)**3) / n) / (std**3)
    return skewness

def ComputeKurtosis(data):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    kurtosis = (np.sum((data - mean)**4) / n) / (std**4) - 3
    return kurtosis
```

## Function to Return Samples based on PDF

```python

from scipy.interpolate import interp1d

def sample_from_pdf(x, pdf, total_samples):
    """
    Generates samples from a given PDF defined on a discrete x-axis using inverse transform sampling.

    Parameters:
    - x (np.ndarray): Points where the PDF is defined (1D array).
    - pdf (np.ndarray): PDF values at corresponding x (must be non-negative).
    - total_samples (int): Number of random samples to generate.

    Returns:
    - samples (np.ndarray): Generated samples of shape (total_samples,)
    """
    # Ensure valid input
    x = np.asarray(x)
    pdf = np.asarray(pdf)
    assert x.ndim == 1 and pdf.ndim == 1 and len(x) == len(pdf), "x and pdf must be 1D arrays of same length"
    assert total_samples > 0, "total_samples must be positive"

    # Normalize PDF to sum to 1
    dx = x[1] - x[0]
    pdf = pdf / np.sum(pdf * dx)

    # Compute CDF
    cdf = np.cumsum(pdf) * dx
    cdf /= cdf[-1]  # normalize CDF to 1

    # Invert CDF
    inv_cdf = interp1d(cdf, x, bounds_error=False, fill_value=(x[0], x[-1]))

    # Generate uniform samples and map to x-space
    u = np.random.rand(total_samples)
    samples = inv_cdf(u)

    return samples
```

# Get Best Estimates of Bin Size - Freedman-Diaconis Rule
##### Using Numpy (Use Numpy as Its faster than Trensorflow Function)

```python
def BinsFreedmanDiaconisRule(X):
    n = len(X)
    iqr = np.percentile(X, 75) - np.percentile(X, 25)
    bin_width = 2 * iqr / n**(1/3)
    return int(np.ceil((np.max(X) - np.min(X)) / bin_width))
```

##### Using Tensorflow - Function to compute Total Bins  (Freedman-Diaconis Rule)

```python
@tf.function(reduce_retracing=True)
def BinsFreedmanDiaconisRuleTF(X):
    X = tf.cast(X, tf.float32)
    n = tf.cast(tf.size(X), tf.float32)

    # Sort X
    X_sorted = tf.sort(X)

    # Find 25th and 75th percentiles manually
    idx_25 = tf.cast(tf.round(0.25 * (n - 1)), tf.int32)
    idx_75 = tf.cast(tf.round(0.75 * (n - 1)), tf.int32)

    q25 = X_sorted[idx_25]
    q75 = X_sorted[idx_75]

    iqr = q75 - q25

    bin_width = 2.0 * iqr / tf.pow(n, 1.0/3.0)
    range_X = tf.reduce_max(X) - tf.reduce_min(X)

    bins = tf.math.ceil(range_X / (bin_width + 1e-6))  # avoid division by zero
    return tf.cast(bins, tf.int32)  
```

## Function To Get x,y Coordinates of PDF from Histogram

```python
def FilteredHistogramPDF(X, bins=50, ax=None, Col='gray',smooth=True, points=500, xmin=None, xmax=None):
    """
    Computes and optionally smooths histogram-based PDF.

    Args:
        X: Input 1D array
        bins: Number of histogram bins
        xmin, xmax: Optional limits to filter the histogram output
        smooth: If True, applies spline smoothing
        points: Number of points for smooth curve
        ax: Matplotlib axes object to plot the histogram (default is None, meaning no plot)
        Col: Color code for the histogram plot (default is 'gray')

    Returns:
        x (array): x values (sorted in ascending order)
        y (array): PDF values (sorted accordingly)
    """
    density=True; #Normalize Histogram - Use Density
    hist, bin_edges = np.histogram(X, bins=bins, density=density)
    x = 0.5 * (bin_edges[:-1] + bin_edges[1:])  # bin centers
    y = hist

    # Filter data within the range [xmin, xmax] if provided
    if xmin is not None and xmax is not None:
        mask = (x > xmin) & (x < xmax)
        x = x[mask]
        y = y[mask]

    # Apply spline smoothing if required
    if smooth and len(x) > 3:
        spline = make_interp_spline(x, y)
        x = np.linspace(x.min(), x.max(), points)
        y = spline(x)

    # Ensure x and y are sorted in ascending x order
    sort_idx = np.argsort(x)
    x = x[sort_idx]
    y = y[sort_idx]

    # Plot histogram on the given axes (if ax is provided)
    if ax is not None:
        ax.hist(X, bins=bins, density=density, alpha=0.2, color=Col, label='Histogram')
    
    return x, y
```

# Asymmetric Laplace PDF

The **Asymmetric Laplace Distribution (ALD)** is a continuous distribution extending the classic Laplace by allowing different decay rates on each side of its central location. Hereâ€™s the mathematical form:

\\[
X \sim \text{AL}(m, \lambda, \kappa),
\\]  
where:

- \( $m$ \): location parameter  
- \( $\lambda > 0 $\): scale  
- \( $\kappa > 0 $\): asymmetry factor

The **probability density function (PDF)** is:

\\[
f(x \mid m, \lambda, \kappa) =
\frac{\lambda}{\kappa + \frac{1}{\kappa}}
\begin{cases}
\exp\left( \frac{\lambda}{\kappa}(x - m) \right), & x < m \\\\
\exp\left( -\lambda \kappa (x - m) \right), & x \ge m
\end{cases}
\\]

When \( $\kappa = 1 $\), this reduces to the symmetric Laplace distribution.

Kotz, S., Kozubowski, T.J., PodgÃ³rski, K. (2001). Asymmetric Laplace Distributions. In: The Laplace Distribution and Generalizations. BirkhÃ¤user, Boston, MA. https://doi.org/10.1007/978-1-4612-0173-1_3


```python
# Define the PDF of the Asymmetric Laplace Distribution
def AsymmetricLaplacePDF(x, m, lambda_, kappa):
    norm_const = lambda_ / (kappa + 1/kappa)
    pdf = np.where(
        x < m,
        norm_const * np.exp((lambda_ / kappa) * (x - m)),
        norm_const * np.exp((-lambda_ * kappa) * (x - m))
    )
    return pdf

# Inverse transform sampling function
def SampleAsymmetricLaplace(m, lambda_, kappa, size):
    u = np.random.uniform(0, 1, size)
    p = kappa**2 / (1 + kappa**2)
    samples = np.where(
        u < p,
        m + kappa / lambda_ * np.log(u / p),
        m - 1 / (lambda_ * kappa) * np.log((1 - u) / (1 - p))
    )
    return samples

if 1: 
    # Parameters for the Asymmetric Laplace Distribution
    m = 0.2       # location
    lambda_ = 1 # scale
    kappa = 0.5 # asymmetry
    N=1000
    

    # Generate x values and corresponding PDF values
    x_vals = np.linspace(m - 3, m + 15, N)
    pdf_vals = AsymmetricLaplacePDF(x_vals, m, lambda_, kappa)

    # Generate the dataset X^
    X_hat = SampleAsymmetricLaplace(m, lambda_, kappa, N)



    # Plot the distribution
    plt.figure(figsize=(8, 4))
    plt.hist(X_hat, bins=100, density=True, alpha=0.6, label="Sampled Histogram")
    plt.plot(x_vals, pdf_vals, label=f"m={m}, Î»={lambda_}, Îº={kappa}")
    plt.title("Asymmetric Laplace Distribution")
    plt.xlabel("x")
    plt.ylabel("PDF")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Compute area under the PDF using Simpson's rule
    area = np.trapz(pdf_vals, x_vals)
    print(area)
```

```python
# Function to compute memory usage in kilobytes (KB)
def ComputeTensorMemory(tensor):
    """
    Compute the memory usage of a tensor in kilobytes (KB).
    Memory usage = Number of elements * Size of each element (in bytes)
    """
    num_elements = tf.size(tensor).numpy()  # Total number of elements in the tensor
    element_size = tensor.dtype.size  # Size of each element in bytes
    memory_usage_kb = num_elements * element_size / 1024  # Convert to KB
    return memory_usage_kb

def ComputeNumpyMemory(array):
    """
    Compute the memory usage of a NumPy array in kilobytes (KB).
    Memory usage = Number of elements * Size of each element (in bytes)
    """
    num_elements = array.size  # Total number of elements
    element_size = array.itemsize  # Size of each element in bytes
    memory_usage_kb = num_elements * element_size / 1024  # Convert bytes to KB
    return memory_usage_kb
```

# Tensorflow Func to compute Prob from Normal Dist estimation

```python
def NormPDF(x, mu, Sigma, RetMem=False):
    # Ensure mu and Sigma are tensors for compatibility with Keras/TensorFlow
    mu = tf.convert_to_tensor(mu, dtype=tf.float32)
    Sigma = tf.convert_to_tensor(Sigma, dtype=tf.float32)

    # Calculate the PDF of a Normal distribution using the formula
    # f(x) = (1 / (Sigma * sqrt(2 * pi))) * exp(-0.5 * ((x - mu) / Sigma)^2)
    
    pi = tf.constant(3.141592653589793, dtype=tf.float32)
    coefficient = 1 / (Sigma * tf.sqrt(2 * pi))
    exponent = -0.5 * tf.square((x - mu) / Sigma)
    y = coefficient * tf.exp(exponent)

    if RetMem:
        # Compute memory utilization in MB
        MemKB = (tf.size(x) + 
                 tf.size(coefficient) + 
                 tf.size(exponent)  + 
                 tf.size(mu) + 
                 tf.size(Sigma)+ 
                 tf.size(y))*4  / (1024)
        return y, MemKB
    else:
        return y
```

```python
if 0:
    # Define Func to return a function RetFunc to compute PDF
    @tf.function(reduce_retracing=True) # important!
    def RetFunc(x, 
                Order, 
                xMid, 
                CoefLeft, 
                CoefRight, 
                MinProb, 
                MaxProb, 
                Auc, 
                RetMem):
        # Ensure the input tensor is of type float32
        x = tf.cast(x, tf.float32)

        # Create the Mask for indices on the left and right of the Peak
        IdxLeft = tf.where(x < xMid, True, False)
        IdxLeft = tf.cast(IdxLeft, dtype=tf.float32)

        IdxRight = tf.where(x >= xMid, True, False)
        IdxRight = tf.cast(IdxRight, dtype=tf.float32)

        # Interpolate the left side values
        xLeft = x * IdxLeft
        yLeft = tf.zeros_like(xLeft, dtype=tf.float32)
        for Power in range(Order):
            yLeft = yLeft + (CoefLeft[Power] * (xLeft ** Power))
        yLeft = yLeft * IdxLeft

        # Interpolate the right side values
        xRight = x * IdxRight
        yRight = tf.zeros_like(xRight, dtype=tf.float32)
        for Power in range(Order):
            yRight = yRight + (CoefRight[Power] * (xRight ** Power))
        yRight = yRight * IdxRight

        # Calculate the probability
        Prob = (yLeft + yRight) / Auc
        # Clip the values < MinProb (Epsilon) & > MaxProb(MaxY)
        Prob = tf.clip_by_value(Prob, clip_value_min=MinProb, clip_value_max=MaxProb)

        # Calculate memory usage if RetMem is True
        if RetMem:
            # Memory calculation (batch_size * 4 * (5 + 2 * Order)) in bytes
            batch_size = tf.size(x)
            mem_bytes = batch_size * 4 * 8 + (2 * Order)
            MemKB = mem_bytes / 1024  # Convert to KB
            return Prob, MemKB
        else: return Prob



    #--------------------------------------------------------
    # Function to Interpolate KDE
    #--------------------------------------------------------
    @tf.function(reduce_retracing=True)
    def IntKDE(X, bandwidth=None, Points=200, RetMem=False):
        X = tf.convert_to_tensor(X, dtype=tf.float32)
        n = tf.cast(tf.shape(X)[0], tf.float32)

        # Use Silverman's Rule of Thumb for bandwidth if not provided
        if bandwidth is None:
            std_dev = tf.math.reduce_std(X)
            bandwidth = 1.06 * std_dev * tf.pow(n, -1.0 / 5)

        # Create sample points over data range
        x_min = tf.reduce_min(X)
        x_max = tf.reduce_max(X)
        x = tf.linspace(x_min, x_max, Points)  # PDF evaluation points

        # Expand for broadcasting
        X_expand = tf.expand_dims(X, axis=1)     # shape (n, 1)
        x_expand = tf.expand_dims(x, axis=0)     # shape (1, m) # m=Points(200)

        # Gaussian KDE: compute kernel values
        diffs = (x_expand - X_expand) / bandwidth   # shape (n, m)
        kernels = tf.exp(-0.5 * tf.square(diffs)) / (bandwidth * tf.sqrt(2.0 * math.pi))  # shape (n, m)

        # Estimate PDF as mean across kernels
        pdf= tf.reduce_mean(kernels, axis=0)     # shape (m,)
        # === Estimate memory ===# Optional
        if RetMem:
            # Compute memory utilization in MB
            Mem_X = tf.cast(n/ (1024**2), tf.float32)
            Mem_x = tf.cast(Points/ (1024**2), tf.float32) 
            Mem_X_Ex = tf.cast(n/ (1024**2), tf.float32)
            Mem_x_Ex = tf.cast(Points/ (1024**2), tf.float32)            
            Mem_diffs = tf.cast(n / (1024 ** 2), tf.float32)* tf.cast(Points, tf.float32)           
            Mem_kernel = tf.cast(n / (1024 ** 2), tf.float32)* tf.cast(Points, tf.float32) 
            Mem_Pdf = tf.cast(Points/ (1024**2), tf.float32) 
            MemMB=(Mem_X+Mem_x+Mem_X_Ex+Mem_x_Ex+Mem_diffs+Mem_kernel+Mem_Pdf)*4  #4 Bytes in tf.float32

            #tf.print("Mem_x: ", Mem_x, " Mem_diffs: ", Mem_diffs, " Mem_kernel: ", Mem_kernel, " Mem_density: ", Mem_density)           
            #tf.print("IntKDE Mem: ", MemMB)
            return x, pdf, MemMB
        else:
            # Optional memory cleanup
            del X, X_expand, x_expand, diffs, kernels
            return x, pdf   
    #************************************************************************    
    # Superset and Subset are 1d tensor
    #@tf.function(reduce_retracing=True)
    def CreateFunc4PDF(Superset,  Points=200, PolyOrder=6, RetProb=False, RetMemKDE=False, RetMem=False, Print=False, USE_MAX_Y=True):

        #--------------------------------------------------------
        Epsilon = tf.constant(1e-6, dtype=tf.float32)
        # Ensure the input tensor is of type float32
        Superset = tf.cast(Superset, tf.float32)
        # Convert it to 1D 
        Superset = tf.reshape(Superset, [-1])
        #----------------------------------------------------------
        if RetMemKDE:
            X, Y, MemMB = IntKDE(Superset, Points=Points, RetMem=RetMemKDE)    
            tf.print("IntKDE Mem: ", MemMB);
        else:
            X, Y = IntKDE(Superset, Points=Points, RetMem=False)  
        #----------------------------------------------------------
        # Get the indices that would sort X
        SortedIdx = tf.argsort(X)
        xSorted = tf.gather(X, SortedIdx)
        ySorted = tf.gather(Y, SortedIdx)

        # Compute the Mid index
        MaxY = tf.reduce_max(ySorted)
        MidIdx = tf.where(ySorted >= MaxY)[0,0] if USE_MAX_Y else TotalPts // 2;
        xAtMid = xSorted[MidIdx]
        yAtMid = ySorted[MidIdx]

        # Create the design matrix for a polynomial of degree = PolyOrder
        xOrder = tf.stack([tf.pow(xSorted, i) for i in range(PolyOrder)], axis=1)
        UpX = xOrder[:MidIdx]
        UpY = ySorted[:MidIdx]
        ExUpY = tf.expand_dims(UpY, axis=1)

        DnX = xOrder[MidIdx:]
        DnY = ySorted[MidIdx:]
        ExDnY = tf.expand_dims(DnY, axis=1)

        # Perform least squares fit For Left Side of Histogram
        CoefUp = tf.linalg.lstsq(UpX, ExUpY)
        # Perform least squares fit for Right Side of Histogram
        CoefDn = tf.linalg.lstsq(DnX, ExDnY)

        # Interpolate new values for the polynomial fit
        xUp = tf.linspace(xSorted[0], xSorted[MidIdx-1], Points)
        xDn = tf.linspace(xSorted[MidIdx], xSorted[-1], Points)

        # Interpolate the left side values
        yUp = tf.zeros_like(xUp, dtype=tf.float32)
        for Power in range(PolyOrder):
            yUp = yUp + (CoefUp[Power] * (xUp ** Power))

        # Interpolate the right side values
        yDn = tf.zeros_like(xDn, dtype=tf.float32)
        for Power in range(PolyOrder):
            yDn = yDn + (CoefDn[Power] * (xDn ** Power))

        # Concatenate the results
        xFit = tf.concat([xDn, xUp], axis=0)
        yFit = tf.concat([yDn, yUp], axis=0)

        # Clip the values < Epsilon & > MaxY
        yFit = tf.clip_by_value(yFit, clip_value_min=Epsilon, clip_value_max=MaxY)

        # Sort the x and y values for plotting
        SortedIdx = tf.argsort(xFit)
        xFitSorted = tf.gather(xFit, SortedIdx)
        yFitSorted = tf.gather(yFit, SortedIdx)

        # Compute the area under the curve (AUC) using the trapezoidal rule
        dx = xFitSorted[1:] - xFitSorted[:-1]
        Av_dy = (yFitSorted[1:] + yFitSorted[:-1]) / 2.0
        BaseAUC = tf.reduce_sum(dx * Av_dy)

        # Always Based on sorted x
        if Print:
            print(f"yAtMidY@Mid: {yAtMid:.1f} MidIdx: {MidIdx} xAtMid: {xAtMid} MaxY: {MaxY:.1f}")  
            print(f"X: {X.shape}-{X.dtype} Y: {Y.shape}-{Y.dtype} xOrder: {xOrder.shape}-{xOrder.dtype}")
            print(f"UpX: {UpX.shape}-{UpX.dtype} UpY: {UpY.shape}-{UpY.dtype} ExUpY: {ExUpY.shape}-{ExUpY.dtype}")
            print(f"DnX: {DnX.shape}-{DnX.dtype} DnY: {DnY.shape}-{DnY.dtype} ExDnY: {ExDnY.shape}-{ExDnY.dtype}")
            print(f"CoefUp: {CoefUp.shape}-{CoefUp.dtype} CoefDn: {CoefUp.shape}-{CoefDn.dtype}")
            print(f"xUp: {xUp.shape}-{xUp.dtype} xDn: {xDn.shape}-{xDn.dtype}")
            print(f"yUp: {yUp.shape}-{yUp.dtype} yDn: {yDn.shape}-{yDn.dtype}")
            print(f"xFit: {xFit.shape}-{xFit.dtype} yFit: {yFit.shape}-{yFit.dtype}")
            print(f"dx: {dx.shape}-{dx.dtype} Av_dy: {Av_dy.shape}-{Av_dy.dtype}")
            print(f"BaseAUC[{BaseAUC.shape}]: {BaseAUC}")

        # Return the probability function
        if RetProb:
            return xFitSorted, yFitSorted
        else:
            # Return a lambda wrapper
            return lambda x: RetFunc(
                x,
                Order=PolyOrder,
                xMid=xAtMid,
                CoefLeft=CoefUp,
                CoefRight=CoefDn,
                MinProb=Epsilon,
                MaxProb=MaxY,
                Auc=BaseAUC,
                RetMem=RetMem
            )        

```

```python
# Define Func to return a function RetFunc to compute PDF
@tf.function(reduce_retracing=True) # important!
def RetFunc(x, 
            Order, 
            xMid, 
            CoefLeft, 
            CoefRight,
            xMin,
            xMax,
            MinProb, 
            MaxProb, 
            Auc, 
            RetMem):
    # Ensure the input tensor is of type float32
    x = tf.cast(x, tf.float32)

    # Create the Mask for indices on the left and right of the Peak
    IdxLeft = tf.logical_and(x >= xMin, x <= xMid)  
    #IdxLeft = tf.where(x < xMid, True, False)
    IdxLeft = tf.cast(IdxLeft, dtype=tf.float32)
    
    IdxRight = tf.logical_and(x >= xMid, x <= xMax) 
    #IdxRight = tf.where(x >= xMid, True, False)
    IdxRight = tf.cast(IdxRight, dtype=tf.float32)

    # Interpolate the left side values
    xLeft = x * IdxLeft
    yLeft = tf.zeros_like(xLeft, dtype=tf.float32)
    for Power in range(Order):
        yLeft = yLeft + (CoefLeft[Power] * (xLeft ** Power))
    yLeft = yLeft * IdxLeft

    # Interpolate the right side values
    xRight = x * IdxRight
    yRight = tf.zeros_like(xRight, dtype=tf.float32)
    for Power in range(Order):
        yRight = yRight + (CoefRight[Power] * (xRight ** Power))
    yRight = yRight * IdxRight

    # Calculate the probability
    Prob = (yLeft + yRight) #/ Auc
    # Clip the values < MinProb (Epsilon) & > MaxProb(MaxY)
    Prob = tf.clip_by_value(Prob, clip_value_min=MinProb, clip_value_max=MaxProb)

    # Calculate memory usage if RetMem is True
    if RetMem:
        # Memory calculation (batch_size * 4 * (5 + 2 * Order)) in bytes
        batch_size = tf.size(x)
        mem_bytes = batch_size * 4 * 8 + (2 * Order)
        MemKB = mem_bytes / 1024  # Convert to KB
        return Prob, MemKB
    else: return Prob



#--------------------------------------------------------
# Function to Interpolate KDE
#--------------------------------------------------------
@tf.function(reduce_retracing=True)
def IntKDE(X, bandwidth=None, Points=200, RetMem=False):
    X = tf.convert_to_tensor(X, dtype=tf.float32)
    n = tf.cast(tf.shape(X)[0], tf.float32)

    # Use Silverman's Rule of Thumb for bandwidth if not provided
    if bandwidth is None:
        std_dev = tf.math.reduce_std(X)
        bandwidth = 1.06 * std_dev * tf.pow(n, -1.0 / 5)

    # Create sample points over data range
    x_min = tf.reduce_min(X)
    x_max = tf.reduce_max(X)
    x = tf.linspace(x_min, x_max, Points)  # PDF evaluation points

    # Expand for broadcasting
    X_expand = tf.expand_dims(X, axis=1)     # shape (n, 1)
    x_expand = tf.expand_dims(x, axis=0)     # shape (1, m) # m=Points(200)

    # Gaussian KDE: compute kernel values
    diffs = (x_expand - X_expand) / bandwidth   # shape (n, m)
    kernels = tf.exp(-0.5 * tf.square(diffs)) / (bandwidth * tf.sqrt(2.0 * math.pi))  # shape (n, m)

    # Estimate PDF as mean across kernels
    pdf= tf.reduce_mean(kernels, axis=0)     # shape (m,)
    # === Estimate memory ===# Optional
    if RetMem:
        # Compute memory utilization in MB
        Mem_X = tf.cast(n/ (1024**2), tf.float32)
        Mem_x = tf.cast(Points/ (1024**2), tf.float32) 
        Mem_X_Ex = tf.cast(n/ (1024**2), tf.float32)
        Mem_x_Ex = tf.cast(Points/ (1024**2), tf.float32)            
        Mem_diffs = tf.cast(n / (1024 ** 2), tf.float32)* tf.cast(Points, tf.float32)           
        Mem_kernel = tf.cast(n / (1024 ** 2), tf.float32)* tf.cast(Points, tf.float32) 
        Mem_Pdf = tf.cast(Points/ (1024**2), tf.float32) 
        MemMB=(Mem_X+Mem_x+Mem_X_Ex+Mem_x_Ex+Mem_diffs+Mem_kernel+Mem_Pdf)*4  #4 Bytes in tf.float32

        #tf.print("Mem_x: ", Mem_x, " Mem_diffs: ", Mem_diffs, " Mem_kernel: ", Mem_kernel, " Mem_density: ", Mem_density)           
        #tf.print("IntKDE Mem: ", MemMB)
        return x, pdf, MemMB
    else:
        # Optional memory cleanup
        del X, X_expand, x_expand, diffs, kernels
        return x, pdf   
#************************************************************************    
# Superset and Subset are 1d tensor
#@tf.function(reduce_retracing=True)
def CreateFunc4PDF(Superset,  Points=500, PolyOrder=6, 
                   xMin= None, xMax= None, TimesStd=3.5,
                   RetProb=False, RetMemKDE=False, RetMem=False, Print=False, USE_MAX_Y=True):
 
    #--------------------------------------------------------
    Epsilon = tf.constant(1e-6, dtype=tf.float32)
    # Ensure the input tensor is of type float32
    Superset = tf.cast(Superset, tf.float32)
    # Convert it to 1D 
    Superset = tf.reshape(Superset, [-1])

    # Compute xMinLeft using mean and std (Outliers)
    if xMin is None:
        xMinLeft = tf.reduce_mean(Superset) - TimesStd * tf.math.reduce_std(Superset)
    else:
        xMinLeft = tf.cast(xMin, tf.float32)

    # Compute xMaxRight using mean and std (outliers)
    if xMax is None:
        xMaxRight = tf.reduce_mean(Superset) + TimesStd * tf.math.reduce_std(Superset)
    else:
        xMaxRight = tf.cast(xMax, tf.float32)    
    tf.print("xMinLeft: ", xMinLeft, " xMaxRight:", xMaxRight)
    #----------------------------------------------------------
    if RetMemKDE:
        X, Y, MemMB = IntKDE(Superset, Points=Points, RetMem=RetMemKDE)    
        tf.print("IntKDE Mem: ", MemMB);
    else:
        X, Y = IntKDE(Superset, Points=Points, RetMem=False)  
    #----------------------------------------------------------
    # Get the indices that would sort X
    SortedIdx = tf.argsort(X)
    xSorted = tf.gather(X, SortedIdx)
    ySorted = tf.gather(Y, SortedIdx)

    # Compute the Mid index
    MaxY = tf.reduce_max(ySorted)
    MidIdx = tf.where(ySorted >= MaxY)[0,0] if USE_MAX_Y else TotalPts // 2;
    xAtMid = xSorted[MidIdx]
    yAtMid = ySorted[MidIdx]

    # Create the design matrix for a polynomial of degree = PolyOrder
    xOrder = tf.stack([tf.pow(xSorted, i) for i in range(PolyOrder)], axis=1)
    UpX = xOrder[:MidIdx]
    UpY = ySorted[:MidIdx]
    ExUpY = tf.expand_dims(UpY, axis=1)

    DnX = xOrder[MidIdx:]
    DnY = ySorted[MidIdx:]
    ExDnY = tf.expand_dims(DnY, axis=1)

    # Perform least squares fit For Left Side of Histogram
    CoefUp = tf.linalg.lstsq(UpX, ExUpY)
    # Perform least squares fit for Right Side of Histogram
    CoefDn = tf.linalg.lstsq(DnX, ExDnY)

    # Interpolate new values for the polynomial fit
    xUp = tf.linspace(xSorted[0], xSorted[MidIdx-1], Points)
    xDn = tf.linspace(xSorted[MidIdx], xSorted[-1], Points)

    # Interpolate the left side values
    yUp = tf.zeros_like(xUp, dtype=tf.float32)
    for Power in range(PolyOrder):
        yUp = yUp + (CoefUp[Power] * (xUp ** Power))

    # Interpolate the right side values
    yDn = tf.zeros_like(xDn, dtype=tf.float32)
    for Power in range(PolyOrder):
        yDn = yDn + (CoefDn[Power] * (xDn ** Power))

    # Concatenate the results
    xFit = tf.concat([xDn, xUp], axis=0)
    yFit = tf.concat([yDn, yUp], axis=0)

    # Clip the values < Epsilon & > MaxY
    yFit = tf.clip_by_value(yFit, clip_value_min=Epsilon, clip_value_max=MaxY)

    # Sort the x and y values for plotting
    SortedIdx = tf.argsort(xFit)
    xFitSorted = tf.gather(xFit, SortedIdx)
    yFitSorted = tf.gather(yFit, SortedIdx)

    # Compute the area under the curve (AUC) using the trapezoidal rule
    dx = xFitSorted[1:] - xFitSorted[:-1]
    Av_dy = (yFitSorted[1:] + yFitSorted[:-1]) / 2.0
    BaseAUC = tf.reduce_sum(dx * Av_dy)

    # Always Based on sorted x
    if Print:
        print(f"yAtMidY@Mid: {yAtMid:.1f} MidIdx: {MidIdx} xAtMid: {xAtMid} MaxY: {MaxY:.1f}")  
        print(f"X: {X.shape}-{X.dtype} Y: {Y.shape}-{Y.dtype} xOrder: {xOrder.shape}-{xOrder.dtype}")
        print(f"UpX: {UpX.shape}-{UpX.dtype} UpY: {UpY.shape}-{UpY.dtype} ExUpY: {ExUpY.shape}-{ExUpY.dtype}")
        print(f"DnX: {DnX.shape}-{DnX.dtype} DnY: {DnY.shape}-{DnY.dtype} ExDnY: {ExDnY.shape}-{ExDnY.dtype}")
        print(f"CoefUp: {CoefUp.shape}-{CoefUp.dtype} CoefDn: {CoefUp.shape}-{CoefDn.dtype}")
        print(f"xUp: {xUp.shape}-{xUp.dtype} xDn: {xDn.shape}-{xDn.dtype}")
        print(f"yUp: {yUp.shape}-{yUp.dtype} yDn: {yDn.shape}-{yDn.dtype}")
        print(f"xFit: {xFit.shape}-{xFit.dtype} yFit: {yFit.shape}-{yFit.dtype}")
        print(f"dx: {dx.shape}-{dx.dtype} Av_dy: {Av_dy.shape}-{Av_dy.dtype}")
        print(f"BaseAUC[{BaseAUC.shape}]: {BaseAUC}")
    
    # Return the probability function
    if RetProb:
        return xFitSorted, yFitSorted
    else:
        # Return a lambda wrapper
        return lambda x: RetFunc(
            x,
            Order=PolyOrder,
            xMid=xAtMid,
            CoefLeft=CoefUp,
            CoefRight=CoefDn,
            xMin=xMinLeft,
            xMax=xMaxRight,
            MinProb=Epsilon,
            MaxProb=MaxY,
            Auc=BaseAUC,
            RetMem=RetMem
        )        

```

# Compute MSE w.r.t PDF Estimation From KDE

```python

def MSE_Compare_PDF(W, y, x, Prob):
    """
    Compute MSE between estimated y values at W and histogram-based PDF (HistProb) defined over x.

    Args:
        W (np.ndarray): Query points where PDF is evaluated.
        y (np.ndarray): PDF values at W (predicted).
        x (np.ndarray): x-values of the histogram PDF.
        Prob (np.ndarray): PDF values at x.

    Returns:
        float: Mean Squared Error
    """
    # Interpolate the PDF at the points in W
    y_Prob = np.interp(W, x, Prob)
    
    # Compute MSE
    mse = np.mean((y - y_Prob) ** 2)
    return mse
```

# Check Pearson distribution types based on skew and kurtosis.

```python
def GetPearsonType(skewness, kurtosis):
    """
    Classifies data into Pearson distribution types based on skewness and kurtosis.
    
    Args:
        skewness (float): Sample skewness.
        kurtosis (float): Sample kurtosis (NOT excess kurtosis).
        
    Returns:
        str: Pearson type.
    """
    # Handle numerical precision issues
    eps = 1e-5
    
    beta1 = skewness**2
    beta2 = kurtosis  # Should be the "normal" kurtosis, not "excess"

    # Calculate kappa (discriminant)
    kappa = beta2 - (3/2) * beta1 - 3

    if np.abs(skewness) < eps:
        # Symmetric cases
        if np.abs(kurtosis - 3) < eps:
            return "Type 0 (Normal distribution)"
        elif kurtosis < 3:
            print("Type II (Symmetric bounded - Beta)")
            return 2
        else:
            print("Type VII (Symmetric heavy tails - t-distribution)")
            return 7
    else:
        # Skewed cases
        if kappa < 0:
            print("Type I (Bounded skewed - Beta)")
            return 1
        elif np.abs(kappa) < eps:
            print("Type III (Unbounded one-tail skewed - Gamma)")
            return 3
        else:
            print("Type IV (General skewed distribution)")
            return 4

```

# Estimates the Pearson Type I PDF

```python
def GetPearsonPara_1(x, skw, kurt):
    """
    Estimates the Pearson Type I distribution parameters (a, b, loc, scale) from data x.
    
    Parameters:
    - x: 1D array-like, input data (sample data).
    
    Returns:
    - a: Shape parameter of the Beta distribution (Pearson Type I)
    - b: Shape parameter of the Beta distribution (Pearson Type I)
    - loc: Location parameter (minimum value of x)
    - scale: Scale parameter (max value - min value of x)
    """
    # Step 1: Estimate loc and scale
    loc = np.min(x)
    scale = np.max(x) - np.min(x)
            
    # Step 2: Estimate a and b based on skewness and kurtosis
    # Using relationships from Pearson Type I distribution
    a = float(2 - (skw/ 2))
    b = float(3 - (kurt/2))
    
    return a, b, loc, scale


def Pearson_I_PDF(x, a, b, loc, scale, RetMem=False):
    """
    Computes the PDF for Pearson Type I distribution.
    
    Parameters:
    - x: 1D array-like or scalar, the points at which to evaluate the PDF.
    - a: Shape parameter of the Beta distribution (positive).
    - b: Shape parameter of the Beta distribution (positive).
    - loc: Location parameter (mean).
    - scale: Scale parameter (standard deviation).
    
    Returns:
    - pdf_vals: The computed PDF values at the input points `x`.
    """
    a=float(a); b=float(b);
    # Transform x to the standard Beta domain [0, 1]
    z = (x - loc) / scale
    
    # Ensure z is in the range [0, 1]
    if np.any(z < 0) or np.any(z > 1):
        return 0.0  # Outside of [loc, loc + scale]
    
    # Calculate the Beta-like distribution's PDF
    norm_const = 1 / (scale * BetaFunc(a, b))  # Correct usage of scipy.special.beta
    pdf_vals = norm_const * z**(a - 1) * (1 - z)**(b - 1)
    if RetMem:
            # Compute memory usage
            x_mem = ComputeTensorMemory(tf.cast(x, tf.float32))#sys.getsizeof(x) / 1024  # Convert bytes to kilobytes
            z_mem = ComputeTensorMemory(tf.cast(z, tf.float32))#sys.getsizeof(z) / 1024
            pdf_vals_mem = ComputeTensorMemory(tf.cast(pdf_vals, tf.float32))#sys.getsizeof(pdf_vals) / 1024
            scalars_mem = 5 * 8 / 1024  # 5 scalars (a, b, loc, scale, norm_const) * 8 bytes each

            total_mem = x_mem + z_mem + pdf_vals_mem + scalars_mem
            return pdf_vals, total_mem
    else: return pdf_vals    


```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the custom StateSpaceLayer (from the previous response)
class StateSpaceLayer(layers.Layer):
    def __init__(self, d_model, N, name=None, **kwargs):
        super(StateSpaceLayer, self).__init__(name=name, **kwargs)
        self.d_model = d_model
        self.N = N
        self.Out=1
    
    def build(self, input_shape):
        # Define state-space model matrices (A, B, C, etc.)
        self.A = self.add_weight(name="A_matrix", shape=(self.d_model, self.d_model), initializer="random_normal")
        self.B = self.add_weight(name="B_matrix", shape=(input_shape[-1], self.d_model), initializer="random_normal")  # Fix: (F, d_model)
        self.C = self.add_weight(name="C_matrix", shape=(self.d_model, self.Out), initializer="random_normal")  # Output to 1
    
    def call(self, inputs):
        # Dynamically get batch size from input tensor shape
        batch_size = tf.shape(inputs)[0]
        
        # Initialize hidden state with zeros
        hidden_state = tf.zeros((batch_size, self.d_model))  # Initialize hidden state

        # Process through each time step (N)
        for t in range(self.N):
            u_t = inputs[:, t, :]  # Feature at time t
            hidden_state = tf.matmul(hidden_state, self.A) + tf.matmul(u_t, self.B)  # Shape: (batch_size, d_model)
        
        print(f"hidden_state: {hidden_state.shape}")
        # Output (predict final sample point)
        output = tf.matmul(hidden_state, self.C)  # Shape: (batch_size, 1)
        return output
    
    def get_config(self):
        config = super(StateSpaceLayer, self).get_config()
        config.update({
            "d_model": self.d_model,
            "N": self.N
        })
        return config
    
    @classmethod
    def from_config(cls, config):
        return cls(**config)    
    
# Test code

# Set parameters for the test
N = 256    # Number of time steps (sample points)
F = 1      # Number of features
d_model = 32  # Size of hidden state vector

# Create a test input tensor of shape (None, N, F)
batch_size = 2  # You can change this to test different batch sizes
test_input = tf.random.normal((batch_size, N, F))

# Define the input layer
inputs = tf.keras.Input(shape=(N, F))
tf.print("input", tf.shape(test_input))
# Create the model with StateSpaceLayer
x = StateSpaceLayer(d_model=d_model, N=N)(inputs)
model = models.Model(inputs, x)

# Summarize the model architecture
model.summary()

# Run the test with the dummy data
output = model(test_input)

# Print the output shape
print("Output shape:", output.shape)
```

```python
import tensorflow as tf
import numpy as np

# Function to create HiPPO matrix (as a TensorFlow constant)
def create_hippo_matrix(N):
    X = np.zeros((N, N), dtype=np.float32)
    for n in range(N):
        for k in range(N):
            if n > k:
                X[n, k] = -np.sqrt(2 * n + 1) * np.sqrt(2 * k + 1)
            elif n == k:
                X[n, k] = -(n + 1)
    return tf.constant(X)

# Function to compute A_bar using TensorFlow operations
def compute_A_bar(A, delta):
    """
    Compute A_bar = (I - (delta/2) * A)^(-1) * (I + (delta/2) * A)
    """
    I = tf.eye(A.shape[0], dtype=tf.float32)  # Identity matrix
    delta_half_A = (delta / 2.0) * A
    I_minus = I - delta_half_A
    I_plus = I + delta_half_A
    A_bar = tf.linalg.inv(I_minus) @ I_plus
    return A_bar

# Set matrix size N
N = 5  # Example size

# Create HiPPO matrix (A) as a TensorFlow constant
A = create_hippo_matrix(N)

# Given delta (step size)
delta = 0.01  # 10ms step size

# Compute A_bar
A_bar = compute_A_bar(A, delta)

# Print both matrices (A and A_bar)
print("HiPPO Matrix A:")
print(A.numpy())  # Convert TensorFlow tensor to NumPy array for printing
print("\nTransformed Matrix A_bar:")
print(A_bar.numpy())  # Convert TensorFlow tensor to NumPy array for printing
```

```python
# Function to create HiPPO matrix using only TensorFlow ops
def create_hippo_matrix_tf(N):
    rows = []
    for n in tf.range(N):
        row = []
        for k in tf.range(N):
            val = tf.cond(n > k,
                lambda: -tf.sqrt(tf.cast(2 * n + 1, tf.float32)) * tf.sqrt(tf.cast(2 * k + 1, tf.float32)),
                lambda: tf.cond(n == k,
                    lambda: -tf.cast(n + 1, tf.float32),
                    lambda: tf.constant(0.0, dtype=tf.float32)))
            row.append(val)
        rows.append(tf.stack(row))
    return tf.stack(rows)

# Function to compute A_bar using TensorFlow operations
def compute_A_bar(A, delta):
    """
    Compute A_bar = (I - (delta/2) * A)^(-1) * (I + (delta/2) * A)
    """
    I = tf.eye(tf.shape(A)[0], dtype=tf.float32)
    delta_half_A = (delta / 2.0) * A
    I_minus = I - delta_half_A
    I_plus = I + delta_half_A
    A_bar = tf.linalg.inv(I_minus) @ I_plus
    return A_bar

# Matrix size and delta
N = 5
delta = 0.01  # For 100Hz sampling

# Generate HiPPO matrix and compute A_bar
A = create_hippo_matrix_tf(N)
A_bar = compute_A_bar(A, delta)

# Run and print matrices in TensorFlow eager mode
tf.print("HiPPO Matrix A:\n", A)
tf.print("\nTransformed Matrix A_bar:\n", A_bar)
```

```python
# Function to compute B_bar = (I - (Î”/2) * A)^(-1) * (Î” * B)
def compute_B_bar(A, B, delta):
    N = tf.shape(A)[0]
    I = tf.eye(N, dtype=tf.float32)
    delta_half_A = (delta / 2.0) * A
    I_minus = I - delta_half_A
    delta_B = delta * B
    B_bar = tf.linalg.inv(I_minus) @ delta_B
    return B_bar

input_dim = 2 # Number of input channels
B = tf.ones((N, input_dim), dtype=tf.float32)  # You can replace with custom B

# Compute B_bar
B_bar = compute_B_bar(A, B, delta)

# Print results
tf.print("Matrix A:\n", A)
tf.print("\nMatrix B:\n", B)
tf.print("\nTransformed Matrix B_bar:\n", B_bar)
```
